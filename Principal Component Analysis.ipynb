# In Depth: Principal Component Analysis
Lets consider unsupervised estimators which can highlight interesting aspects of data without reference to any known labels

PCA is broadly used and is a dimensionality reduction algorithm
- can ve useful for visulization, noise filtering, feature extraction

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

## Introducing Principal Component Analysis
Consider a 2d dataset with 200 points

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
```

We can see a linear relationship between x and y

However instead of predicting the value of y, we want to learn about the relationship between x and y

For PCA, this relationship is quantified by finding a list of the principal axes in the data, and using them to describe the dataset

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

The first learns components and explained variance

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

to understand them, lets visualize them as input data:
    - components as the direction vector
    - explained variance as squared-length of the vector

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

These vecotrs represent the principal axes of the data
- length of vector indicates how important that axis is for the distribution of the data
- it measures variance of the data projected on that axis
- projection of each data point onto the principal axes are the principal components

This transformation is an affine transformation

## PCA as dimensionality reduction
- zero out 1 or more of the smallest principal components
- this results in a lower dimensional projection of the data that preserves the maximal data variance

```python
# dimensionality reduction transform
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print('original shape: ', X.shape)
print('transformed shape: ', X_pca.shape)
```

To understand the effect of this, let's inverse transform the data annd plot it

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal')
```

The reduced dataset has lost variance in the smaller principal axis, but it retains the overall replathionship

## PCA for visualization: Hand-written digits
the usefulness of this is more evident in higher dimensions.

Consider hand written digits

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

Use PCA to project this 64d data into a more managable number of dimensions

```python
pca = PCA(2)
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

Plot t he first 2 principal components of each point to learn about the data

```python
plt.scatter(projected[:, 0], projected[:, 1],
           c=digits.target, edgecolor='none', alpha=0.5,
           cmap=plt.cm.get_cmap('spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

The points of the projection of an image along directions with largest variance
- optimal stretch and rotation

## What do the components mean?
The reduced dimensions are combinations of basis vectors

It allows for low dimensional representation of the data
- it is more efficent that the native pixel-basis of the input data

## Choosing the number of components
It is important to estimate how many components are neededto describe the data
- We can look at the cumulative explained variance ratio as a function of the number of components

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

We can see that 10 components can describe 75% of the data
- 50 components could describe 100% of the data

## PCA as Noise Filtering
components with variance much larger than the effect of the noise should be relatively unaffected by the noise.
- we can reconstruct the largest subset of the principal components
- this should keep the signal and remove the noise

```python
# plot some clean digits
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                            subplot_kw={'xticks':[], 'yticks':[]},
                            gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                 cmap='binary', interpolation='nearest',
                 clim=(0, 16))
plot_digits(digits.data)        
```

```python
# add some random noise
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy)
```

```python
# train PCA on the noisy data, then request the projection to preserve 50% of the variance
pca = PCA(0.5).fit(noisy)
pca.n_components_
```

Here, 50% variance is 12 principal components.

- compute these components and use the inverse transform to reconstruct the filtered digits

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

This property is very useful. You might train on lower dimensional data, then use it to automatically filter out random noise on inputs

## Example: Eigenfaces
Explore PCA projection as a feature selector for facial recognition

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

Look at the principal axes that span this dataset
- it is large so we will use the RandomizedPCA
- this randomizes the approximate first N principal components

Look at the frist 150 components

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(150)
pca.fit(faces.data)
```

Visualize the first several principal components
- known as eigenvectors

```python
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                        subplot_kw={'xticks': [], 'yticks': []},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

This gives insight to show images vary.
The eigenfaces seem to be picking out certain features

Lets look at the cumulative variance of these components to see how much data information the projection is preserving

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

These 150 components account for just over 90% variance. 

Reconstruct the images and compare

```python
# compute the components and projected faces
pca = RandomizedPCA(150).fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)
```

```python
# Plot the results
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');
```
