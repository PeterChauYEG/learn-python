# In Depth: Linear Regression
- a very good starting point for regression tasks
- can be fit very quickly and are very interpretable
- the simplist form is a linear regerssion
- can be extended to model more complicated data behaviour

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

## Simple linear regression
to fit data to a straight line, we look at the equation: $y = ax + b$

`a` is the slope and `b` is the intercept

Lets consider this data which is scattered about a line with slope 2, and an intercept of -5

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y)
```

```python
# use the linearRegression estimator to fit this data and construct a line of best fit
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
```

```python
# cehck the model's fit params
print("Model slope:      ", model.coef_[0])
print("Model intercept:  ", model.intercept_)
```

The results are very close to the actual line

This estimator can also handle multidimensional linear models of the form:
    $y = a0 + a1x1 + a2x2 + ...$
    
This is akin to fitting a plane to points in multiple dimensions.

```python
rng = np.random.RandomState(10)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])

model.fit(X, y)
print(model.intercept_)
print(model.coef_)
```

This works well for linear relationships, but we can also relex these restrictions

## Basis Function Regression
- We can adapt linear regressions to nonlinear relationships between variables by transforming data according to basis functions.
- The idea is to take a multidimensional linear model, and project it into a higher dimension

### Polynomial basis functions

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

- we've convertedd the 1d array into a 3d array where each column is exponentally larger
- this new higher-dimensional representation can be plugged into a linear regression

```python
# use a pipeline to create a 7th degree polynomial modal
from sklearn.pipeline import make_pipeline
poly_model = make_pipeline(PolynomialFeatures(7), LinearRegression())
```

```python
# now that we've transformed the data, we can fit it to see the more complicated relathionship
# lets use a sine wave with noise
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

poly_model.fit(x[:, np.newaxis], y)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
```

This 7th order basis function usage makes a linear regression work for even sine waves!

### Gaussian Basis Functions
- we can also use the sum of gaussian bases
- lets create a customer transformer

```python
from sklearn.base import BaseEstimator, TransformerMixin

class GaussianFeatures(BaseEstimator, TransformerMixin):
    """Uniformly spaced Gaussian features for 1d input"""
    
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
    
    @staticmethod
    def _gauss_basis(x, y, width, axis=None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
    
    def fit(self, X, y=None):
        # create N centers spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
    
    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_, self.width_,
                                axis=1)
    
gauss_model = make_pipeline(GaussianFeatures(20), LinearRegression())
gauss_model.fit(x[:, np.newaxis], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10)
```

## Regularization
- basis functions makes the model more flexible, but can lead to overfitting.

```python
model = make_pipeline(GaussianFeatures(30), LinearRegression())
model.fit(x[:, np.newaxis], y)

plt.scatter(x, y)
plt.plot(xfit, model.predict(xfit[:, np.newaxis]))

plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)
```

This is way overfit. Lets plot the coefficients of the Gaussian bases with respect to their locations

```python
def basis_plot(model, title=None):
    fig, ax = plt.subplots(2, sharex=True)
    model.fit(x[:, np.newaxis], y)
    ax[0].scatter(x, y)
    ax[0].plot(xfit, model.predict(xfit[:, np.newaxis]))
    ax[0].set(xlabel='x', ylabel='y', ylim=(-1.5, 1.5))
    
    if title:
        ax[0].set_title(title)
        
    ax[1].plot(model.steps[0][1].centers_,
              model.steps[1][1].coef_)
    ax[1].set(xlabel='basis location',
              ylabel='coefficient',
              xlim=(0, 10))
    
model = make_pipeline(GaussianFeatures(30), LinearRegression())
basis_plot(model)
```

The amplitude of the basis function at each location is shown.
- when there is overfit from basis function overlap, the amplitudes blow up and cancel each other out.
- we can limit the spikes and penalize large values of the model parameters
- this is known as regularization

### Ridge Regression ($L_2$ regularization)
- the most common form.
- penalizing the sum of squares (2-norms) of the model coefficients

```python
from sklearn.linear_model import Ridge
model = make_pipeline(GaussianFeatures(30), Ridge(alpha=0.1))
basis_plot(model, title='Ridge Regression')
```

the alpha parameter is the knob controlling the complexity of the resulting model.
- 0 is a linear regression
- as it approaches infinity, the model responses is surpressed.

This is useful as it is also computationally efficient

### Lasso Regression ($L_1$ regularization)
- we can also penalize by the sum of absolute values of regression coeffients
- tends to favor sparse models

```python
from sklearn.linear_model import Lasso
model = make_pipeline(GaussianFeatures(30), Lasso(alpha=0.001))
basis_plot(model, title="Lasso Regression")
```

The majority of the coefficients are exactly zero.

## Example: Predicting Bicycle Traffic

Lets Predict the number of bicycle trips across Seattle's Fremont Bridge based on weather, season, and other factors
- join bike data with another dataset
- try to determine the extent to which weather and seasonal factors - temp, precipitation, and daylight hours affect the volumn of bicycle traffic through this corridor

NOOA makes their faily weather station data available
- join the datasets

Perform a simple linear regression to relate the weather and other info to bicycle counts
- estimate how a change in these affects the number of riders on a given day

```python
# join datasets, index by date
# !curl -o FremontBridge.csv https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD
import pandas as pd
counts = pd.read_csv('data/FremontBridge.csv', index_col='Date', parse_dates=True)
weather = pd.read_csv('data/BicycleWeather.csv', index_col='DATE', parse_dates=True)
```

```python
# Compute total daily bicycle traffic and put it in a dataframe
daily = counts.resample('d').sum()
daily['Total'] = daily.sum(axis=1)
daily = daily[['Total']] # remove other columns
```

```python
# we previously saw that the patterns vary day to day. Account for this by adding binary columns that indicate the day of the week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
    daily[days[i]] = (daily.index.dayofweek ==i).astype(float)
```

```python
# We might expect riders to behave differently on holidays. Add this too
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays('2012', '2016')

daily = daily.join(pd.Series(1, index=holidays, name='holiday'))
daily['holiday'].fillna(0, inplace=True)
daily.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# hours of daylight may also affect this. Lets use standard astronomical calucations to add this information
def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given date"""
    days = (date - pd.datetime(2000, 12, 21)).days
    m = (1. - np.tan(np.radians(latitude))
         * np.tan(np.radians(axis) * np.cos(days * 2 * np.pi / 365.25)))
    return 24. * np.degrees(np.arccos(1 - np.clip(m, 0, 2))) / 180.
```

```python
daily['daylight_hrs'] = list(map(hours_of_daylight, daily.index))
daily[['daylight_hrs']].plot()
plt.ylim(8, 17)
```

```python
# Add the average temp and total precipitation to the data
# also add a flag for if the day is dry (zero precipitation)
# temperatures are in 1/10 deg C: convert to C
weather['TMIN'] /= 10
weather['TMAX'] /= 10
weather['Temp (C)'] = 0.5 * (weather['TMIN'] + weather['TMAX'])

# precip is in 1/10 mm; convert to inches
weather['PRCP'] /= 254
weather['dry day'] = (weather['PRCP'] == 0).astype(int)

daily = daily.join(weather[['PRCP', 'Temp (C)', 'dry day']])
daily.head()
```

```python
# add a counter that increases from day 1 that means number of years that have past
# allows use to see if there is any observed increase or decrease
daily['annual'] = (daily.index - daily.index[0]).days / 365.
```

```python
# lets look at the dataset we built
daily.head()
```

```python
# Lets fit this linear regression
# use fit_intercept=False cause the daily flags operate as their own day-specific intercepts
from sklearn.linear_model import LinearRegression

# drop any rows with null values
daily.dropna(axis=0, how='any', inplace=True)

column_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'holiday', 'daylight_hrs', 'PRCP', 'dry day', 'Temp (C)', 'annual']
X = daily[column_names]
y = daily['Total']

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
daily['predicted'] = model.predict(X)
```

```python
# look at data and compare predictions to actual
daily.head()
daily[['Total', 'predicted']].plot(alpha=0.5)
```

We've missed some features in the summer. They might not be complete, but this isn't a bad model

Look at the coefficients to estimate how much each feature contributes to the daily bicycle count

```python
params = pd.Series(model.coef_, index=X.columns)
params
```

```python
# Lets compute uncertainties with bootstrap resampling of the data
from sklearn.utils import resample
np.random.seed(1)
err = np.std([model.fit(*resample(X, y)).coef_
             for i in range(1000)], 0)
```

```python
# Look at coeffients with errors
print(pd.DataFrame({'effect': params.round(0),
                   'error': err.round(0)}))
```

We can see that for each additional hour of daylight, 129 +- 9 more people ride.

We also threw away fine grained details. There are also nonlinear relationships
