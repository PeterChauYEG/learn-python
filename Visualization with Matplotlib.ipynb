## Visualizing Errors
- for any scientific measurement, accurate accounting for errors is nearly as imporant as the number itself.

### Basic Errorbars

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
# fmt controls the type of errorbar
# we may want to use a lighter bar for tighter data
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
            ecolor='lightgray', elinewidth=3, capsize=0)
```

### Continuous Errors
- we can use plt.plot and plt.fill_between to create this
- lets persom a Gaussian process regression
    - a method of fitting a very flexible non-parametric function to data with a continuous measure of uncertainty

```python
from sklearn.gaussian_process import GaussianProcess

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# compute the Gaussian process fit
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,
                    random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE) # 2*sigma ~ 95% configence region
```

```python
# we now have xfit, yfit, and dygit which sample the continuous fit to our data
# instead of ploting 1000 points' errorberas, lets use fill_between to visualize the continuous error
```

```python
# visulize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                color='gray', alpha=0.2)
plt.xlim(0, 10)
```

```python
# this shows what the Gaussian process regression algorithm does.
# we fill where the x value with y-lower and y-upper
```

## Density and Contour Plots
- sometimes it is useful to display 3d data in 2d using contours or color-coded regions
- we can use plt.contour for contour plots
- plot.contourf for filled contour plots
- plt.imshow for showing images

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### Visualizing a 3d function
- lets demo the contour plot using the function $z = f(x, y)$

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

Contour plots can be created with the plt.contour function
- it takes 3 values:
    - x, y, z
    - z is the contour level
 
 np.meshgrid can be used to build 2d grids from 1d arrays

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# line-only contour plot
# by default, negative values get dashed lines
plt.contour(X, Y, Z, colors='black')
```

```python
# lines can be color-coded by specifying a colormap with the cmap arg
# lets also use more lines, 20 equally spaced intervals within the data range
# use the Red-Gray color map
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

```python
# lets fill the contours
# also add a colorbar for scale
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
# the graph is splotchy cause the color is discrete rather than continuous.
# use imshow to render the data as an image
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

- imshow doesn't accept an x and y grid so we need to specify the extent.
- most contour plot's origin is in the lower left, imshow origin is the upper left
- it also automatically adjusts the axis aspect ratio to match the input data

```python
# lets combine a contour plot and an image plot 
# with a transparent background image, and the contour having labels
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

- to create a histogram in 2d we divide the number line into 2d bins

### Kernal density estimation
- a common method for evaluating densities in multiple dimensions 
- a way to smear out the points in space and add up the result to obtain a smooth function

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# plot the result of as an image
plt.imshow(Z.reshape(Xgrid.shape),
          origin='lower', aspect='auto',
          extent=[-3.5, 3.5, -6, 6],
          cmap='Blues')
cb = plt.colorbar()
cb.set_label('density')
```

```python
# KDE has a smoothing length that effectively slides the knob between detail and smoothness
```

### Multiple legends
- we can't just add another legend
- we need to make one via ax.add_artist()

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                    styles[i], color='black')
ax.axis('equal')

# specify the lines and labes of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
         loc='upper right', frameon=False)

# Create the second legend and add the artist manually
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
            loc='lower right', frameon=False)
ax.add_artist(leg)
```

## Choosing the colormap
- there are 3 categories of colormaps:
    - sequential colormaps: a continuous sequence of colors (binary or viridis)
    - divergent colormaps: 2 distinct colors that show positive and negative deviations from a mean (RdBu or PuOr)
    - qualitative colormaps: mix colors with no particular sequence (rainbow or jet)
        
qualitative colormaps are a bad choice for quantitative maps because there is no uniform progression in brightness as the scale increases

```python
from matplotlib.colors import LinearSegmentedColormap

def grayscale_cmap(cmap):
    """Return a grayscale version of the given colormap"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    # convert RGBA to perceived grayscale luminance
    # cf. http://alienryderflex.com/hsp.html
    RGB_weight = [0.299, 0.587, 0.114]
    luminance = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]
        
    return LinearSegmentedColormap.from_list(cmap.name + "_gray", colors, cmap.N)
    

def view_colormap(cmap):
    """Plot a colormap with its grayscale equivalent"""
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(2, figsize=(6, 2),
                           subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([colors], extent=[0, 10, 0, 1])
    ax[1].imshow([grayscale], extent=[0, 10, 0, 1])
```

```python
view_colormap('jet')
```

```python
# viridis is constructed to have an even brightness variation across the range
view_colormap('viridis')
```

```python
# for rainbow schemes, cubehelix is a good colormap
view_colormap('cubehelix')
```

```python
# RdBu is good for showing positive and negative deviations
view_colormap('RdBu')
```

### Example: handwritten digits
- lets use data of thumbails of various handwritten digits
- download them and visualize several of them

```python
# load images of the digits 0 through 5 and visualize several of them
from sklearn.datasets import load_digits
digits = load_digits(n_class=6)

fig, ax = plt.subplots(8, 8, figsize=(6, 6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
# each digit is defined by the hue of its 64 pixels
# each digit is a point lying ins 64 dimensional space
# each dimension is the brightness of one pixel
# visualizing high-dimensional spaces can be difficult
# we can use dimensionality reduction techniques such as manifold learning to reduce the dimensionality of the data while maintaining relationships of interest
# this is an example of machine learning
# lets look at 2d manifold learning projection of this digits data
```

```python
# project the digits into 2 dimensions using IsoMap
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)

# define a discrete colormap
# set ticks and climt
# plot the results
plt.scatter(projection[:, 0], projection[:, 1], lw=0.1,
           c=digits.target, cmap=plt.cm.get_cmap('cubehelix', 6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)
```

```python
# the projection shows some interesting relationships
# 5 and 3 overlap meaning they are difficult to distingush
# 0 and 1 don't overlap as they are very different
```

## Text and Annotation
- useful for convaying interestin information

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-white')
import numpy as np
import pandas as pd
```

### Example: effect of holidays on US births
- lets work with the birthrate data again

```python
births = pd.read_csv('data/births.csv')

quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')

births['day'] = births['day'].astype(int)

births.index = pd.to_datetime(10000 * births.year +
                             100 * births.month +
                             births.day, format='%Y%m%d')
births_by_date = births.pivot_table('births',
                                   [births.index.month, births.index.day])
births_by_date.index = [pd.datetime(2012, month, day)
                       for (month, day) in births_by_date.index]
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)
```

```python
# we can annotate with plt.text / ax.text at an x/y
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

# add labels to the plot
style = dict(size=10, color='gray')

ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas", ha='right', **style)

# label the axes
ax.set(title='USA births by day of the year (1969-1988)',
      ylabel='average daily births')

# format the x axis with centered month labels
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'))
```

### Transforms and Text position
- previously we anchored text to data locations
- we may want to anchor text to a position on the axes or figure
- we use transforms to do this
-there are 3 pre-defined transforms:
    - ax.transData: to data coordindates
    - ax.transAxes: to axes
    - ax.transFigure: to figure

```python
fig, ax = plt.subplots(facecolor='lightgray')
ax.axis([0, 10, 0, 10])

# transform=ax.transData is the default
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
# transData gives the usual data coordinates
# transAxes gives the location from the bottom-left corner o the axes as a fraction of the axes size
# transFigure is similar but from the bottom-left of the figure
```

```python
# notice how changes to the limits effects some transforms
ax.set_xlim(0, 2)
ax.set_ylim(-6, 6)
fig
```

### Fancy tick formats

```python
# plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

# set up grid, legend and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi)
```

```python
# we can add locators with multiplelocator
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
```

```python
# we can format these new locators
# make it more readable
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
    
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```
