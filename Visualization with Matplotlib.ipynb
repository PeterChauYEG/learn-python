### plotting from a script
- in a script, use plt.show()
- it starts the event loop to display the figure
- should be run once per session

```python
# file: myplot.py
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

### Two interfaces for the price of one
- offers 2 interfaces: MATLAB-style, and OOP

#### MATLAB-style interface

```python
plt.figure() # create plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # rows, columns, panel number
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# note that this is stateful
```

#### Object oriented interface 
- offers more control

```python
# first create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# call plot method on appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## Simple Line Plots
- the simplest of all plot visualizations of a single function

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# start by creating a figure and axes
fig = plt.figure()
ax = plt.axes()
```

```python
# figure is a container of all objects...axes graphics text labels
# graphy some data. 
# lets do a sinusoid
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
# we can use the pylab interface and let the figure and axes be created for us
plt.plot(x, np.sin(x))
```

```python
# to createa single figure with multiple lines, call the plot function multiple times
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### Adjusting the plot: line colors and styles
- we can adjust the line color by giving an additional argument to plt.plot()

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g') # short color code
plt.plot(x, np.sin(x - 2), color='0.75') #greyscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # hex code
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) #RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse') # html colors
```

```python
# line style can be adjusted
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# short hand codes
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':') # dotted
```

### Adjusting the plot: axes limits
- we can adjust limits with xlim and ylim

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# plt.axis() allows you to set both
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# we can automatically tighten bounds
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
# we can set units equal to aspact ratios
plt.plot(x, np.sin(x))
plt.axis('equal')
```

### Labeling plots

```python
plt.plot(x, np.sin(x))
plt.title('A sine curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# we can also assign labels
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
# ax.set() allows us to set many props at once
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
      xlabel='x', ylabel='sin(x)',
      title='A Simple Plot')
```
