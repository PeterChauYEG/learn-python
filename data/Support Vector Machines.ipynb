## Motivating Support Vector Machines
- discriminative classification
- finding a line or curve (2d) or manifold (multiple dimensions) that divides classes from each other

Consider a simple classification task where 2 classes of points are well seperated

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                 random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
```

Although we can draw a line to seperate this data by hand, there are more than 1 possibilities

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')
    
plt.xlim(-1, 3.5)
```

Athought these lines seperate the classes perfectly, a new data point will have a different class based on which line you choose.

We need to think deeper
