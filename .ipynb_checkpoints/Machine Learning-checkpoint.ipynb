## What is Machine Learning
- building models of data
- we want to build a mathematical model to help understand data
- learning is the tuning of parameters that can be adapted to observed data
- the fitted model can then be used to predict new data

### Categories of Machine Learning
- there are 2 types:
    - supervised learning
    - unsupervised learning
    
- supervised: 
    - modeling relationship of labeled data and measured features
    - which we then use to apply labels to new data
    - subdivided into classification and regression tasks
        - classification: labels are discrete categories
        - regression: labels are continuous quantities
- unsupervised:
    - modeling features of a dataset without reference to any label
    - task include clustering and dimensionality reduction
        - clustering identifies distinct groups of data
        - dimensionality reduction searchs for more succinct representations

### Supervised Learning Example: Simple Linear Regression
consider fitting a line to $(x, y)$

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y)
```

#### 1. Choose a class of model
- This looks like a linear regression

```python
from sklearn.linear_model import LinearRegression
```

#### 2. Choose model hyperparameters
We may need to answer more questions:
    - would we like to fit for the offset
    - would we like the model to be normalized
    - would we like to preprocess our features to add more model flexibility
    - what degree of regularization would we like to use in our model
    - how many model components would we like to use

For our linear regress, lets fit the intercept

```python
model = LinearRegression(fit_intercept=True)
model
```

#### 3. arrange data
- we need to convert our x into [n_samples, n_features]

```python
x.shape
```

```python
X = x[:, np.newaxis]
X.shape
```

#### 4. Fit the model

```python
model.fit(X, y)
```

```python
# model params get undercores
model.coef_
```

```python
model.intercept_
```

#### 5. Predict labels for unknown data
- generate some new data
- ask what y should be

```python
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
# visualize the results
# plot raw data, then model fit
plt.scatter(x, y)
plt.plot(xfit, yfit)
```

#### Supervised learning example: Iris classification
- Given a model trained on a porton of the Iris data, how well can we predict the remaining labels?

- We will use a simple generative model called Gaussian naive Bayes
- assumes each class is drawn from an axis-aligned Gaussian distribution
- this has no hyperparameters to choose
- good for basline classification

- we want to evaluate the data so we need to split it into training data and testing data

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB() # 2. instantiate model
model.fit(Xtrain, ytrain) # 3. fit model to data
y_model = model.predict(Xtest) # 4. preict on new data
```

```python
# check fraction of predicted labels that match their true value
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

#### Unsupervised learning example: Iris dimensionality
- lets reduce the dimensionality of the Iris data so that it's easier to visualize
- lets use principal component analysis
- it is a fast linear dimensionality reduction technique
- ask to return 2 components

```python
from sklearn.decomposition import PCA # 1. choose the model class
model = PCA(n_components=2) # 2. instantiate the model
model.fit(X_iris) # 3. fit the data
X_2D = model.transform(X_iris) # 4. transform the data into 2 dimensions
```

```python
# plots the results by inserting them into the original iris dataframa
# lets use lmplot
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot('PCA1', 'PCA2', hue='species', data=iris, fit_reg=False)
```

this worked pretty well!

#### Unsupervised Learning: iris clustering
- lets apply clustering to the iris data
- this will attempt to find distinct groups of data without reference to any labels
- lets try Gaussian mixture model
- it attempts to model data as a collection of gaussian blobs

```python
from sklearn.mixture import GMM # 1. choose model class
model = GMM(n_components=3, covariance_type='full') # 2. instantiate the model
model.fit(X_iris) # 3. fit to data
y_gmm = model.predict(X_iris) # 4. determine cluster labels
```

```python
# add cluster to dataframe and plot
iris['cluster'] = y_gmm
sns.lmplot('PCA1', 'PCA2', data=iris, hue='species', col='cluster', fit_reg=False)
```

```python
applying GMM to the PCA data pulls apart species very well
```

## Hyperparameters and Model Validation
- choosing the class of model and hyperparameters may be the most important part of using these tools
- we need to validate both to get the best results

### Thinking about model validation
- the act of estimating how effective it is by applying training data and comparing it to a predicted known value
- there are naive ways of doing this
- we need to use holdout sets and cross-validation for more robust model evaluation

### Modal validation the wrong way
- lets look at a naive approach

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

Choose a model and hyper parameters
- Lets use k-neighbors classifier and n_neighbors=1
- this will label the unknown point the same as the closest training point

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
```

Train the model and predict labels for data we already know

```python
model.fit(X, y)
y_model = model.predict(X)
```

compute the fraction of correctly labeled points

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_model)
```

This is inaccurate because we are testing it with data it's already seen.

### Model validation the right way: Holdout sets
- it would be better to hold on to some data that the model is never trained on. Then use it to test the mdodel

```python
from sklearn.cross_validation import train_test_split
# split the data with 50% in each set
X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.5)

# fit the model on 1 set of data
model.fit(X1, y1)

# evaluate the model on the other set of data
y2_model = model.predict(X2)
accuracy_score(y2, y2_model)
```

This is much more reasonable. The holdout set is also similar to the training set

### Model validation via cross-validation
- 1 disadvantage of a holdout set is that we lost data for training.
- this is not optimal if the training set is small
- we can use cross-validation to address this
- this means to do a sequence of fits where each subset of data is used both as a training set and as a validation set

We will do 2 validation trials. Using a different half as the holdout set

```python
y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

Now we have 2 accuracy scores, which when combined give a better measure of global performance
- what we did was a 2-fold cross-validation

We can also do this for more trials and more folds such as a 5-fold cross-validation
- split the data into 5 groups
- use them to fit on the other 4/5 of the data
- we can use cross_val_score to do this

```python
from sklearn.cross_validation import cross_val_score
cross_val_score(model, X, y, cv=5)
```

Repeating the validation across different subsets gives an even better idea of the performance of the algorithm

We can also train on all data except 1 point
- leave-one-out cross validation

```python
from sklearn.cross_validation import LeaveOneOut
scores = cross_val_score(model, X, y, cv=LeaveOneOut(len(X)))
scores
```

In this case, 1 means successful, and 0 means unsuccessful.

Lets take the mean to get the error rate

```python
scores.mean()
```

### Selecting the best model
If our estimator is underperforming, how should we move forward?
- use a more complicated / more flexible model
- use a less complicated / less flexible model
- gather more training samples
- grather more data to add features to each sample

#### Bias-variance trade-off
the tradeoff is often between bias and variance
- high bias will underfit
    - model can't handle all features of the data
- high variance will overfit
    - model fitted too well to all features..  even noise

Look at the coefficient of determination, a measure of how well a model perfroms relative to a mean of the target values
- $R^2$

$R^2=1$ indicates a perfect match
$R^2=0$ indicates the model does no better than taking the mean of the data
$R^2 < 0$ indicates the model is worst

From these scores, we can make the observation that:
    - high bais performance on the validation set is similar to performance on the training set
    - high variance performance on the validation set is worst than performance on the traning set

Validation Curve:
    - training score is higher than the validation score
        - model better fir to data it has seen than that it has not
    - a low model complexity (high bais), training data is underfit
        - model is a poor predictor both for trianing and unseen data
    - high model complexity is overfit
        - model predicts training data well but fails for unseen data
    - some intermediate value validation curve has a maximum
        - indicates best tradeoff
        
#### Validation curves in scikit-learn
- lets use cross-validation to compute the validation curve for a class of models
- we will use a polynomial regression model
- implement a simple linear regression with a polynominal preprocessor
- use a pipeline to string these operations together

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                        LinearRegression(**kwargs))
```

```python
# create data to fit the model
import numpy as np

def make_data(N, err=1.0, rseed=1):
    # randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y

X, y = make_data(40)
```

```python
# visualize data and polynomial fit of several degrees
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

X_test = np.linspace(-0.1, 1.1, 500)[:, None]

plt.scatter(X.ravel(), y, color='black')
axis = plt.axis()
for degree in [1, 3, 5]:
    y_test = PolynominalRegression(degree).fit(X, y).predict(X_test)
    plt.plot(X_test.ravel(), y_test, label='degree={0}'.format(degree))
plt.xlim(-0.1, 1.0)
plt.ylim(-2, 12)
plt.legend(loc='best')
```

The control for model complexity here is the degree of the polynomial. What degree provides a suitable tradeoff between bias and variance?
We can visualize the validation curve for data and a model
- give a model, data, parameter name, and range to explore
- validation_curve will compute training score and validation scores

```python
from sklearn.learning_curve import validation_curve
degree = np.arange(0, 21)
train_score, val_score = validation_curve(PolynomialRegression(), X, y,
                                         'polynomialfeatures__degree', degree, cv=7)
plt.plot(degree, np.median(train_score, 1), color='blue', label='training score')
plt.plot(degree, np.median(val_score, 1), color='red', label='validation score')
plt.legend(loc='best')
plt.ylim(0, 1)
plt.xlabel('degree')
plt.ylabel('score')
```

- As expected, the training score is higher than validation score
- training score imrpoves with increasing model complexity
- validation score reaches a maximum before dropping off as the model become overfit

the curve shows that the optimal tradeoff is with a third-order polynomial

```python
# compute and display
plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = PolynomialRegression(3).fit(X, y).predict(X_test)
plt.plot(X_test.ravel(), y_test)
plt.axis(lim)
```

We dont need an exact training score since we know it's the best for this type of model from the validation curve

### Learning Curves
- an optimal model depends on the size of training data
- lets examine a new data set with a factor of 5 more points

```python
X2, y2 = make_data(200)
plt.scatter(X2.ravel(), y2)
```

Now lets plot the validation curve for this larger data set. We will also over-plot the previous results

```python
degree = np.arange(21)
train_score2, val_score2 = validation_curve(PolynomialRegression(), X2, y2,
                                           'polynomialfeatures__degree', degree, cv=7)

plt.plot(degree, np.median(train_score2, 1), color='blue', label='training score')
plt.plot(degree, np.median(val_score2, 1), color='red', label='validation score')
plt.plot(degree, np.median(train_score, 1), color='blue', alpha=0.3, linestyle='dashed')
plt.plot(degree, np.median(val_score, 1), color='red', alpha=0.3, linestyle='dashed')
plt.legend(loc='lower center')
plt.ylim(0, 1)
plt.xlabel('degree')
plt.ylabel('score')
```

This new dataset can support a much more complicated model
- it peaks around degree of 6

Validation curves have 2 important inputs:
    - model complexity
    - number of training points
    
It is useful to explore the behavior of the model as a function of the number of training points
- use an increasingly larger subset of data
- a plot of training /validation score with respect to size of the training set is the learning curve

in general, learning curves:
    - model of given complexity will overfit a small dataset
        - training score will be high, validation score will be low
    - model of given complexity will undersit a large dataset
        - training score will decrease, validation score will increase
    - curve should never cross
    - should converge to a particular score as number of training samples grows
        - adding more data will not help
        
### Learning curves
- compute the learning curve for a 2nd order polynomial model and a 9th order

```python
from sklearn.learning_curve import learning_curve

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for i, degree in enumerate([2, 9]):
    N, train_lc, val_lc = learning_curve(PolynomialRegression(degree),
                                        X, y, cv=7,
                                        train_sizes=np.linspace(0.3, 1, 25))
    
    ax[i].plot(N, np.mean(train_lc, 1), color='blue', label='training score')
    ax[i].plot(N, np.mean(val_lc, 1), color='red', label='validation score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1]]), N[0], N[-1],
                color='gray', linestyle='dashed')
    
    ax[i].set_ylim(0, 1)
    ax[i].set_xlim(N[0], N[-1])
    ax[i].set_xlabel('training size')
    ax[i].set_ylabel('score')
    ax[i].set_title('degree = {0}'.format(degree), size=14)
    ax[i].legend(loc='best')
```

This shows how the model responds to increasing training data

### Validation in practice: grid search
in practice, models have many tunable parameters. This makes visualizations difficult.
- We want to find the model that maximizes the validation scores

Lets use grid search to find the optimal polynomial model. 
- explore a 3d grid of model features
    - polynomial degree
    - flag to indict fit of intercept
    - flag to indict wheather to normalize the problem

```python
from sklearn.grid_search import GridSearchCV

param_grid = {'polynomialfeatures__degree': np.arange(21),
             'linearregression__fit_intercept': [True, False],
             'linearregression__normalize': [True, False]}

grid = GridSearchCV(PolynomialRegression(), param_grid, cv=7)
```

```python
# now fit it to data
grid.fit(X, y)
```

```python
# ask for best params
grid.best_params_
```

```python
# visualize it
model = grid.best_estimator_

plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = model.fit(X, y).predict(X_test)
plt.plot(X_test.ravel(), y_test, hold=True)
plt.axis(lim)
```

We can also specify a custom scoring function, parrallize the computations, and do random searches
