They seek to learn:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

## Introducing k-Means
- searches for a pre-determined number of clusters within an unlabeled multidimensional dataset
- it uses a simple conception of what the optimal cluster looks like
    - cluster center is the arithmetic mean of all the points belonging to the cluster
    - each point is closer to it's own cluster center than other cluster centers
    

Examine a simple dataset and see the k-means result.
- generate a 2dd dataset containing four distinct blobs

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

We can pick out clusters by eye. k-means can do this automatically

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Visualize results by plotting data colors by labels.
- plot the cluster centers too

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

k-means invloves an intuitive iterative approach known as expectation-maximization

## k-Means algorithm: expectation-maximization
- a powerful algorithm for a variety of data science contexts

Consists of:
    - guess some cluster centers
    - repeat until converged
        - E-step: assign points to the nearest cluster center
        - M-step: set the cluster centers to the mean
            
- expectation step involves updating our expectatiion of which cluster each point belongs to
- maximization step involves maximising a fitness function that defines the location of cluster centers

Consider this k-means algorithm implementation

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

## Caveats of expectation-maximization
### globally optimal result may not be achieved
- although the result will improve with each step, there is no guarantee of a global best.
- changing random seed will effect the result

```python
centers, labels = find_clusters(X, 4, rseed=0)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

it is common to run the algorithm from mltiple starting guesses
- scikit-learn does this by default

### the number of clusters must be selected before hand

```python
labels = KMeans(6, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

### k-means is limited to linear cluster boundaries
the fundamental assumption means that this algorithm is ineffective if the clusters have complicated geometris.
- boundaries are always linear

```python
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=0.05, random_state=0)

labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

We can use kernel transformation to project the data into high dimensions to discover nonlinear boundaries.

the Spectral Clustering estimator is a kernelized k-means.
- it graphs the nearest neighbors to compute a higher-dimensional representation of the data, then assigns labels using k-means

```python
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis')
```

### k-means can be slow for a large number of samples
accessing every point means the algorithm is slower as the dataset grows.

We can relax this requirement of using all data at every iteration.
- we can use a subset to update the cluster centers at each step
- this is the idea for batched-based k-means

## Examples
### k-means on digits
- attempt to use k-means to id similar digits without using the original label
- load the digits, then find kmeans clusters

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

now cluster

```python
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

The 10 centers took on 64-d. Lets visualize them

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

kmeans is able to find clusters that center on digits

Lets match each learned cluster label with the true label

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

Now check how accurate our unsupervized clustering was in finding similar digits within the data

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

Now use a confusion matrix

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
           xticklabels=digits.target_names,
           yticklabels=digits.target_names)
plt.xlabel('true labels')
plt.ylabel('predicted labels')
```

while there is some confusion, the classifier works pretty well for having now labels.

Lets use the t-distributed stochastic neighbor embedding algorithm to preprocess the data before performing k-means.
- t-SNE is a nonlinear embedding algorithm that is adept at preserving points within clusters

```python
from sklearn.manifold import TSNE

# project the data
tsne = TSNE(n_components=2, init='random', random_state=0)
digits_proj = tsne.fit_transform(digits.data)

# compute the clusters
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits_proj)

# permute the labels
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
    
# compute the accuracy
accuracy_score(digits.target, labels)
```

Sick. We were able to get to nearly 92%
