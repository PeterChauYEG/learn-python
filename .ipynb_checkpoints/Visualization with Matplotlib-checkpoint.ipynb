### Two interfaces for the price of one
- offers 2 interfaces: MATLAB-style, and OOP

#### MATLAB-style interface

```python
plt.figure() # create plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # rows, columns, panel number
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# note that this is stateful
```

#### Object oriented interface 
- offers more control

```python
# first create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# call plot method on appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## Simple Line Plots

### Adjusting the plot: line colors and styles
- we can adjust the line color by giving an additional argument to plt.plot()

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g') # short color code
plt.plot(x, np.sin(x - 2), color='0.75') #greyscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # hex code
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) #RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse') # html colors
```

```python
# line style can be adjusted
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# short hand codes
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':') # dotted
```

## Simple Scatter Plots
- instead of points being joined by line segments, they are just points

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
```

```python
# the 3rd arg is the type of symbol to be used
# there are many possibilities
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
            label="marker='{0}".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

```python
# we can even make a scatter plot with lines
plt.plot(x, y, '-ok')
```

```python
# there are many properties as well
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

### Scatter plots with plt.scatter
- each point's properties can be manipulated

```python
plt.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar()
```

```python
# colors are automatically mapped to the colorbar
# sizes are given in pixels
# these can be used to convay information
```

```python
# lets visualize the size and color of iris flowers
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
# for large datasets, plot is more efficent than scatter
```

## Visualizing Errors
- for any scientific measurement, accurate accounting for errors is nearly as imporant as the number itself.

### Basic Errorbars

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
# fmt controls the type of errorbar
# we may want to use a lighter bar for tighter data
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
            ecolor='lightgray', elinewidth=3, capsize=0)
```

### Continuous Errors
- we can use plt.plot and plt.fill_between to create this
- lets persom a Gaussian process regression
    - a method of fitting a very flexible non-parametric function to data with a continuous measure of uncertainty

```python
from sklearn.gaussian_process import GaussianProcess

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# compute the Gaussian process fit
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,
                    random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE) # 2*sigma ~ 95% configence region
```

```python
# we now have xfit, yfit, and dygit which sample the continuous fit to our data
# instead of ploting 1000 points' errorberas, lets use fill_between to visualize the continuous error
```

```python
# visulize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                color='gray', alpha=0.2)
plt.xlim(0, 10)
```

```python
# this shows what the Gaussian process regression algorithm does.
# we fill where the x value with y-lower and y-upper
```

## Density and Contour Plots
- sometimes it is useful to display 3d data in 2d using contours or color-coded regions
- we can use plt.contour for contour plots
- plot.contourf for filled contour plots
- plt.imshow for showing images

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### Visualizing a 3d function
- lets demo the contour plot using the function $z = f(x, y)$

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

Contour plots can be created with the plt.contour function
- it takes 3 values:
    - x, y, z
    - z is the contour level
 
 np.meshgrid can be used to build 2d grids from 1d arrays

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# line-only contour plot
# by default, negative values get dashed lines
plt.contour(X, Y, Z, colors='black')
```

```python
# lines can be color-coded by specifying a colormap with the cmap arg
# lets also use more lines, 20 equally spaced intervals within the data range
# use the Red-Gray color map
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

```python
# lets fill the contours
# also add a colorbar for scale
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
# the graph is splotchy cause the color is discrete rather than continuous.
# use imshow to render the data as an image
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

- imshow doesn't accept an x and y grid so we need to specify the extent.
- most contour plot's origin is in the lower left, imshow origin is the upper left
- it also automatically adjusts the axis aspect ratio to match the input data

```python
# lets combine a contour plot and an image plot 
# with a transparent background image, and the contour having labels
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

## Histograms, Binnings, and Density
- a simple histogram can be a great first step in understanding a dataset

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data)
```

```python
# there are many customization options
plt.hist(data, bins=30, normed=True, alpha=0.5,
        histtype='stepfilled', color='steelblue',
        edgecolor='none')
```

```python
# using histype='stepfilled', and some transparency alphas is useful for comparing several distributions
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# we can compute the histogram without displaying it
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

### Two-dimensional Histograms and binnings
- to create a histogram in 1d we divide the number line into bins
- to create a histogram in 2d we divide the number line into 2d bins

```python
# define data, x and y array from a multivariate Gaussian distribution
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

#### plt.hist2d: 2d histogram

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

#### plt.hexbins: hexagonal binnings
- creates tesselation of squares across the axes
- another shape is the hexagon

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

### Kernal density estimation
- a common method for evaluating densities in multiple dimensions 
- a way to smear out the points in space and add up the result to obtain a smooth function

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# plot the result of as an image
plt.imshow(Z.reshape(Xgrid.shape),
          origin='lower', aspect='auto',
          extent=[-3.5, 3.5, -6, 6],
          cmap='Blues')
cb = plt.colorbar()
cb.set_label('density')
```

```python
# KDE has a smoothing length that effectively slides the knob between detail and smoothness
```

## Customizing Plot legends
- Gives meaning to a visualization
- plt.legend() automatically creates a legend for any labeled plot elements

```python
import matplotlib.pyplot as plt
plt.style.use('classic')

%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend()
```

```python
# specify location and turn off the frame
ax.legend(loc='upper left', frameon=False)
fig
```

```python
# we can use a fancy box and style it
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```

### Choosign elements for the legend
- we can specify which lines to show in the legend
- plt.plot() returns lines
- passing these to plt.legend and labels will render just those

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

# lines is a list of plt.Line2D instances
plt.legend(lines[:2], ['first', 'second'])
```

### Legend for size of points
- we may want to label the size of points to mark certain features
- lets look at the size of points to indicate populations of cali cities
- the legend will specify the size of the points
- we do this by plotting some labeled data with no entries

```python
import pandas as pd
cities = pd.read_csv('data/california_cities.csv')

# extract the data we're interested in
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

# scatter the points, using size and color but no label
plt.scatter(lon, lat, label=None,
           c=np.log10(population), cmap='viridis',
           s=area, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# createa a legend:
# plot empty list with desired size and label
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
               label=str(area) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area')
plt.title('California Cities: Area and Population')
```

### Multiple legends
- we can't just add another legend
- we need to make one via ax.add_artist()

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                    styles[i], color='black')
ax.axis('equal')

# specify the lines and labes of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
         loc='upper right', frameon=False)

# Create the second legend and add the artist manually
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
            loc='lower right', frameon=False)
ax.add_artist(leg)
```
