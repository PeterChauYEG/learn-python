# In-depth: Support Vector Machines
A powerful and flexible class of supervised algorithms for classification and regression.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

## Motivating Support Vector Machines
- discriminative classification
- finding a line or curve (2d) or manifold (multiple dimensions) that divides classes from each other

Consider a simple classification task where 2 classes of points are well seperated

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                 random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
```

Although we can draw a line to seperate this data by hand, there are more than 1 possibilities

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')
    
plt.xlim(-1, 3.5)
```

Athought these lines seperate the classes perfectly, a new data point will have a different class based on which line you choose.

We need to think deeper

## Support Vector Machines: Maximizing the Margin
Instead of drawing a zero-width line, lets draw a line with some margin. 
- up to the nearest point

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                    color='#AAAAAA', alpha=0.4)
    
plt.xlim(-1, 3.5)
```

The line that maximizes the margin will be chosen as the `maximum margin estimator`

## Fitting a support vector machine
Examine a fit with a linear kernal and a learn c number

```python
from sklearn.svm import SVC # suppport vector classifier
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

```python
# Visualize this with decision boundaries
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
              levels=[-1, 0, 1], alpha=0.5,
              linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                  s=300, linewidth=1, facecolors='none')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model)
```

the dividing line maximizes the margin between the two sets of points.
- the training points just touch the margin
- these points are known as support vectors
- they are stored in `support_vectors_`

```python
model.support_vectors_
```

```python
Only the position of the support vectors matter, any other points don't matter

We can see this by looking a larger amounts  of data`m
```

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2, random_state=0, cluster_std=0.60)
    
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)
    
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

Even though we doubled the amount of data given, the lines don't change. This insensitivity give the SVM model it's strength

```python
from ipywidgets import interact, fixed
interact(plot_svm, N=[10, 200], ax=fixed(None))
```

## Beyond linear boundaries: Kernel SVM
- SVM becomes extremely powerful when combined with kernels
- this is motivated by data that is not linearly separable

```python
from sklearn.datasets.samples_generator import make_circles
X, y = make_circles(100, factor=.1, noise=.1)

clf = SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf, plot_support=False)
```

No linear discrimination will ever be able to seperate this data
- we can project the data into a high dimension so that a linear seperator would be sufficient
- we can use a radial basis function centered on the middle clump

```python
r = np.exp(-(X ** 2).sum(1))
```

Visualize this extra data dimension using a 3d plot

```python
from mpl_toolkits import mplot3d
%matplotlib inline

def plot_3D(elev=30, azim=30, X=X, y=y):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=50, cmap='autumn')
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')
    
interact(plot_3D, elev=[-90, 90], azip=(-180, 180),
        X=fixed(X), y=fixed(y))
```

With the additional dimension, the data is seperable
- centering the radial basis function and tuning the project is important
- we need to automatically find the best basis functions to use

We can compute a basis function centered at every point and let SVM sift through the results
- Kernal transformation
- based on similarity relationship between each pair of points

This my become computationally intensive

We can use the Kernal Trick
- fit on kernel-transformed data can be done implicitly, without building the full represenation of the kernel projection
- it is built into SVM

```python
clf = SVC(kernel='rbf', C=1E6)
clf.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=300, lw=1, facecolors='none')
```

This kernel transformation strategy furns linear methods into fast nonlinear methods

## Tuning the SVM: Softening Margins
- What if the data has some overlap?

```python
X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=1.2)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
```

SVM provides a fudge-factor that softens the margins. This allows some points to creep into the margin
- Large C = hard bargin
- small C = soft margin

```python
X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=0.8)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel='linear', C=C).fit(X, y)
    axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
               model.support_vectors_[:, 1],
               s=300, lw=1, facecolors='none')
    axi.set_title('C = {0:.1f}'.format(C), size=14)
    
```

To determine the optimal value of C, we should use cross-validation

## Example: Face Recognition
Consider the facial recognition problem.
- use the Labeled Faces in the Wild dataset
- thousands of collated photos of various public figures

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
# Plot some of the faces
fig, ax = plt.subplots(3, 5)
for i, axi in enumerate(ax.flat):
    axi.imshow(faces.images[i], cmap='bone')
    axi.set(xticks=[], yticks=[], xlabel=faces.target_names[faces.target[i]])
```

Each image is [62x47] = ~3000 pixels
- instead of using every pixel value as a feature, lets use principal component analysis to extract 150 fundamental components
- then feed it into a SVM classifier
- build a pipeline to do this

```python
from sklearn.svm import SVC
from sklearn.decomposition import RandomizedPCA
from sklearn.pipeline import make_pipeline

pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

```python
# split the dataset
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(faces.data, faces.target, random_state=42)
```

Use grid search cross-validation to explore combinations of parameters
- adjust C (margin hardness) and gamma (size of radial basis function kernel)

```python
from sklearn.grid_search import GridSearchCV
param_grid = {'svc__C': [0.01, 0.1, 1, 5, 10, 50],
              'svc__gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05]}
grid = GridSearchCV(model, param_grid)

%time grid.fit(Xtrain, ytrain)
print(grid.best_params_)
```

```python
# use the best esitmator 
model = grid.best_estimator_
yfit = model.predict(Xtest)
```

```python
# Plot some test images with predicted values
fig, ax = plt.subplots(4, 6)
for i, axi in enumerate(ax.flat):
    axi.imshow(Xtest[i].reshape(62, 47), cmap='bone')
    axi.set(xticks=[], yticks=[])
    axi.set_ylabel(faces.target_names[yfit[i]].split()[-1],
                  color='black' if yfit[i] == ytest[i] else 'red')
fig.suptitle('Predicted Names: Incorrect Labels in Red', size=14)
```

```python
# get a list of recovery statistics label by label
from sklearn.metrics import classification_report
print(classification_report(ytest, yfit, target_names=faces.target_names))
```

```python
# display a confusion matrics between classes
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, yfit)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=faces.target_names,
            yticklabels=faces.target_names)
plt.xlabel('true labels')
plt.ylabel('predicted labels')
```

This gives us a sense of which labels are likely to be confused.

In the real world, photos do not come precropped into nice grids.
- we need to find faces and extract features that are independent of the pixellation

## Support Vector Machine Summary
Powerful because:
- dependant on few support vectors that result in compact models
- predictions is fast
- works well in high-dimensions
- kernel methods are very versatile

Disadvantages:
- large numbers of training samples can be computationally prohibitive
- cross-validation for C is expensive as datasets grow
- results don't have direct probabilistic interpretation
